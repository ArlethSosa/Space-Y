### **Tratamiento de datos nulos**
Identificar los valores nulos para tratarlos

```python
df.isnull().sum()
```

Tenemos 26 datos  de Landing Pad con datos nulos y 5 de payload Mass.

Para tratar los datos de Payload Mass, se calcula el promedio y remplazamos los valores nulos por el promedio

Por otra parte, para Landing Pad los valores nulos representan que no se utilizó una plataforma de aterrizaje por lo que se van a remplazar por "Sin plataforma"

```python
 # Caluclar primedio de la columna PayloadMAss

mean = df["PayloadMass"].mean()

# Remplazar datos

df["PayloadMass"]= df["PayloadMass"].replace(np.nan, mean)
```

```python
df["LandingPad"] = df["LandingPad"].replace(np.nan, "without landing pad")
```

Verificamos que ya no hay valores nulos

```python
df.isnull().sum()
```

Creamos una nueva columan landing_class categorizando los valores de outcome de los aterrizajes

 0 para atterizajes fallidos

 1 para aterrizajes exitosos


```python
landing_outcomes = df["Outcome"].value_counts()
landing_outcomes
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome)
```

```python
""" Creamos una nueva columan landing_class categorizando los valores de outcome de los aterrizajes
 0 para atterizajes fallidos
 1 para aterrizajes exitosos
 """

landing_class=[]
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])

for i in df["Outcome"]:
  if i in bad_outcomes:
      landing_class.append(0)
  else:
      landing_class.append(1)
landing_class
```

```python
# Agregamos la columna al dataframe
df['Class']=landing_class
df[['Class']].head(8)
```
