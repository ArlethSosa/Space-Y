```python
#Lista los nombres de las series de Falcon 9  que tuvieron éxito en el aterrizaje en el barco en el océano y que tienen una masa de carga útil mayor que 4000 pero menor que 6000
%%sql
SELECT "Serial"
FROM SPACEXTABLE
WHERE "Outcome" = 'True Ocean'
AND "PayloadMass" BETWEEN 4000 AND 6000;
```

```python
#Lista de posibles outcome

%sql select distinct "Outcome" from SPACEXTABLE
```

```python
#Mosta la fecha del primer lanzamiento con aterrizaje exitoso en Oceano, ASDS, RTLS
%%sql
SELECT
    "Outcome",
    MIN("Date") AS first_success_date
FROM SPACEXTABLE
WHERE "Outcome" LIKE 'True%'
GROUP BY "Outcome";
```

**2.** Numero de veces a los que fue lanzado un coehete a cada orbita

```python
# Numero de lanzamientos por orbita a la que se lanzaron sin contar GTO porque es solo estacionariaria
df[df['Orbit'] != 'GTO']['Orbit'].value_counts()
```

- LEO (Low Earth Orbit / Órbita terrestre baja):órbita cercana a la Tierra (hasta ~2,000 km), usada para satélites, observación y la ISS.

- VLEO (Very Low Earth Orbit / Órbita terrestre muy baja): Órbita aún más cercana (menos de ~450 km), usada para imágenes de muy alta resolución.

- GTO (Geostationary Transfer Orbit / Órbita de transferencia geoestacionaria): órbita intermedia entre LEO y GEO, se usa para llevar satélites a órbita geoestacionaria.

- SSO (Sun-Synchronous Orbit / Órbita heliosíncrona): Órbita polar que mantiene la misma hora solar al pasar, usada para satélites de clima y observación.

- ES-L1 (Punto Lagrange L1 Tierra-Sol): Punto donde se equilibran las fuerzas Tierra-Sol, permite observar el Sol de forma constante.

- HEO (Highly Elliptical Orbit / Órbita altamente elíptica): Órbita alargada alrededor de la Tierra.

- ISS (International Space Station / Estación Espacial Internacional): Estación habitable en órbita baja, usada para investigación científica.

- MEO (Medium Earth Orbit / Órbita media terrestre): Órbita entre LEO y GEO, usada para GPS y navegación.

- GEO (Geostationary Orbit / Órbita geoestacionaria): A ~35,786 km sobre el ecuador, usada para telecomunicaciones.

- PO (Polar Orbit / Órbita polar): Órbita que pasa por los polos.


**3.** Calcular frecuencia de exito o error en el resultado de la misión (aterrizaje)

```python
landing_outcomes = df["Outcome"].value_counts()
landing_outcomes
```

- True Ocean: aterrizaje exitoso en una zona del océano
- False Ocean: aterrizaje fallido en el océano
- True RTLS: aterrizaje exitoso en plataforma terrestre
- False RTLS: aterrizaje fallido en plataforma terrestre
- True ASDS: aterrizaje exitoso en barco dron (drone ship)
- False ASDS: aterrizaje fallido en barco dron
- None ASDS / None None: fallo en el aterrizaje

EL aterrizaje con más frecuencia es exitoso en barco dron, seguido de los aterrizajes fallidos.
