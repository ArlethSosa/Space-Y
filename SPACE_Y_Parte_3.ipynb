# Ejercicios de uso de SQL en python

```python
# 1. Mostrar los nombres de los lugares de lanzamientos sin repeticiones

%sql select distinct "LaunchSite" from SPACEXTABLE
```

```python
# 2. Mostrar 5 registros donde los sitios de lanzamiento comienzan con la cadena "CCS"


%sql select * from SPACEXTABLE Where "LaunchSite" like "CCS%" limit 5
```

```python
#Lista los nombres de los propulsores (boosters) que tuvieron éxito en el aterrizaje en el barco en el océano y que tienen una masa de carga útil mayor que 4000 pero menor que 6000
%%sql
SELECT "BoosterVersion"
FROM SPACEXTABLE
WHERE "Outcome" = 'True Ocean'
AND "PayloadMass" BETWEEN 4000 AND 6000;
```

```python
%sql select distinct "Outcome" from SPACEXTABLE
```

```python
#Mosta la fecha del primer lanzamiento con aterrizaje exitoso en Oceano, ASDS, RTLS
%%sql
SELECT
    "Outcome",
    MIN("Date") AS first_success_date
FROM SPACEXTABLE
WHERE "Outcome" LIKE 'True%'
GROUP BY "Outcome";
```
