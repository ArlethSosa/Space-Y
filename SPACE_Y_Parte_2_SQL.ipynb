# Ejercicios de uso de SQL en python

```python
# 1. Mostrar los nombres de los lugares de lanzamientos sin repeticione s

%sql select distinct "LaunchSite" from SPACEXTABLE
```

```python
# 2. Display 5 records where launch sites begin with the string 'CCS'¶

%sql select * from SPACEXTABLE Where "LaunchSite" like "CCS%" limit 5
```

```python
#List the names of the boosters which have success in ocean ship and have payload mass greater than 4000 but less than 6000
%%sql
SELECT "BoosterVersion"
FROM SPACEXTABLE
WHERE "Outcome" = 'True Ocean'
AND "PayloadMass" BETWEEN 4000 AND 6000;
```

```python
%sql select distinct "Outcome" from SPACEXTABLE
```

```python
#Mosta la fecha del primer lanzamiento con aterrizaje exitoso en Oceano, ASDS, RTLS
%%sql
SELECT
    "Outcome",
    MIN("Date") AS first_success_date
FROM SPACEXTABLE
WHERE "Outcome" LIKE 'True%'
GROUP BY "Outcome";
```
